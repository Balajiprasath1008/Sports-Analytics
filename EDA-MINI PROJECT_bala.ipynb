### 3.	The following columns need to be converted for further analysis:

```python
l=[]
for i in df['Value'].values:
    i=str(i).replace('€','')
    if 'M' in i:
        a=i.replace('M','')
        b=float(a)*1000000
        l.append(b)     
    elif 'K' in i:
        c=i.replace('K','')
        d=float(c)*1000
        l.append(d) 
    else:
        l.append(i)
df['Value']=l
```

```python
df['Value']=df['Value'].astype('float')
```

```python
df['Value'].head()
```

```python
l=[]
for i in df['Wage'].values:
    i=str(i).replace('€','')
    if 'M' in i:
        a=i.replace('M','')
        b=float(a)*1000000
        l.append(b)     
    elif 'K' in i:
        c=i.replace('K','')
        d=float(c)*1000
        l.append(d) 
    else:
        l.append(i)
df['Wage']=l
```

```python
df['Wage']=df['Wage'].astype('float')
```

```python
df['Wage'].head()
```

```python
df['Joined']=pd.DatetimeIndex(df['Joined']).year
```

```python
df['Joined']=df['Joined'].astype('int')
df['Joined']
```

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'])
```

```python
df['Contract Valid Until'].head()
```

```python
df['Height'].head()
```

```python
l=[]
for i in df['Height']:
        a=str(i).replace("'",".")
        l.append(float(a))
df['Height']=l
df['Height']
```

```python
df['Weight']=df['Weight'].str.replace('lbs','')
df['Weight']=df['Weight'].astype('float')
```

```python
df['Weight']
```

```python
l=[]
for i in df['Release Clause'].values:
    i=str(i).replace('€','')
    if 'M' in str(i):
        a=i.replace('M','')
        b=float(a)*1000000
        l.append(b)     
    elif 'K' in str(i):
        c=i.replace('K','')
        d=float(c)*1000
        l.append(d) 
    else:
        l.append(i)
df['Release Clause']=l
```

```python
df['Release Clause']=df['Release Clause'].astype('float')
```

```python
df['Release Clause'].head()
```

```python
df.shape
```

```python
df.dropna(subset=['Club'],inplace=True)
```

```python
df.dropna(subset=['Preferred Foot'],inplace=True)
```
